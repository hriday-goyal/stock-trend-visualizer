```python
df = pd.read_csv('stock_data_processed.csv', index_col=0)
df = df[['Close']]
df.dropna(inplace=True)
df.head()
```

```python
def create_lag_features(data, n_lags=5):
    df_lagged = data.copy()
    for i in range(1, n_lags + 1):
        df_lagged[f'lag_{i}'] = df_lagged['Close'].shift(i)
    df_lagged.dropna(inplace=True)
    return df_lagged

df_lagged = create_lag_features(df)
df_lagged.head()
```

```python
X = df_lagged.drop('Close', axis=1)
X = df_lagged.drop('Close', axis=1)

# Convert all lag columns to numeric (coerce errors to NaN)
X = X.apply(pd.to_numeric, errors='coerce')

# Drop any rows with NaNs (just in case)
X = X.dropna()

# Match y to the same filtered rows
y = y.loc[X.index]

y = df_lagged['Close']

# Remove non-numeric data
X = X.select_dtypes(include=['float64', 'int64'])

# Train/test split
split_index = int(0.8 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(X_train.shape, y_train.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

# Align y_test to the same length as y_pred
y_test_aligned = y_test.iloc[:len(y_pred)]
```

```python
mae = mean_absolute_error(y_test_aligned, y_pred)
rmse = np.sqrt(mean_squared_error(y_test_aligned, y_pred))

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
```
